```python
# Now let's load read a dataframe from a csv.
# Go to https://www.kaggle.com/datasets/rounakbanik/pokemon/ and download the dataset
# Read that dataset. How you do this depends on where you're running your notebook. 
# If you're running this locally, you'll specify the filepath. 
# If you're running this in kaggle, upload the file and use the filepath that's generated.
full_pokemon_df = pd.read_csv("/kaggle/input/pokemon/pokemon.csv")
full_pokemon_df
```

```python
# Let's get just the useful columns
df = full_pokemon_df[['pokedex_number', 'name', 'generation', 'type1', 'type2', 'attack', 'defense', 'speed', 'sp_attack', 'sp_defense', 'hp']]
df
```

```python
# Practice filtering 
df[(df['generation'] == 1) & (df['attack'] > 50)]
```

```python
# Sort the dataframe
df.sort_values(by=['generation', 'name'], ascending=[True, False])
```

```python
# Practice filtering - way 1
grass_pokemon_df = df[(df['type1'] == 'grass') | (df['type2'] == 'grass')]

# Practice filtering - way 2
grass_pokemon_df = df[df.eval("type1 == 'grass' or type2 == 'grass'")]
```

```python
# Perform numerical caluclations, examples:
df['attack'].mean()
df[['attack', 'defense']].mean()
df.mean(numeric_only=True)
df.describe()
```

```python
# Create a df with the 2 type columns combined:
type1_df = df.drop(columns=['type2']).rename(columns={'type1': 'type'})
type2_df = df.drop(columns=['type1']).rename(columns={'type2': 'type'})
combined_type_df = pd.concat([type1_df, type2_df]).sort_values(by='pokedex_number')
combined_type_df = combined_type_df.dropna()
combined_type_df
```

```python
combined_type_df.groupby(['generation', 'type']).mean(numeric_only=True)
```

```python
# Now let's read another dataframe so we can join tables. This one is from 
# https://github.com/veekun/pokedex/blob/master/pokedex/data/csv/pokemon_evolution.csv
# Same as above, the way you read this depends on where you're running the code.
evolutions_df = pd.read_csv("/kaggle/input/evolutions2/pokemon_evolution (1).csv")
evolutions_df
```

```python
evolutions_df.columns
```

```python
# Merge
df.merge(evolutions_df, left_on=["pokedex_number"], right_on=["id"])

# You can also do inner/outer/left/right/etc
df.merge(evolutions_df, left_on=["pokedex_number"], right_on=["id"], how="left")
```

```python
# Let's see only pokemon that need an item to evolve
df.merge(evolutions_df, left_on=["pokedex_number"], right_on=["id"])[[
    'pokedex_number', 'name', 'generation', 'minimum_level', 'held_item_id'
]].dropna(subset=["held_item_id"])
```

```python
# Let's see the average evolution level by type
combined_type_df.merge(evolutions_df, left_on=["pokedex_number"], right_on=["id"])[[
    'pokedex_number', 'type', 'minimum_level'
]].dropna(subset=["minimum_level"]).groupby('type')[['minimum_level']].mean(numeric_only=True)
```

```python
# Speed test
import time
full_pokemon_df_xl = pd.concat([full_pokemon_df] * 500)

#The slow way
def sum(row):
    return row['attack'] + row['defense'] + row['speed']
    
start = time.time()
full_pokemon_df_xl['total_score_slow'] = full_pokemon_df_xl.apply(sum, axis=1)
end = time.time()
print(end-start)

# The fast way
start = time.time()
full_pokemon_df_xl['total_score_fast'] = full_pokemon_df_xl['attack'] + full_pokemon_df_xl['defense'] + full_pokemon_df_xl['speed']
end = time.time()
print(end-start)
```
